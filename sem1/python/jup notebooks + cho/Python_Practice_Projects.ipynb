# 📝 Python Practice Projects

This notebook contains 3 real-world demo projects for practicing:
- Loops (`for`, `while`)
- Decision statements (`if/elif/else`)
- Functions (`def`, return values)

Projects:
1. ATM Simulation
2. Student Grading System
3. Shopping Cart Billing System


## 🔹 Project 1: ATM Simulation

**Problem Statement:**
Simulate a simple ATM machine where a user can check balance, deposit money, withdraw money, or exit.

**Input Format:**
- User chooses from menu options (1-4).
- For deposit/withdraw, user enters an amount.

**Output Format:**
- Display balance after each transaction.
- Show error if withdrawal amount is greater than balance.


**Practice Exercise:**
- Add a PIN system before showing the menu.
- Limit the number of wrong attempts to 3.


## 🔹 Project 2: Student Grading System

**Problem Statement:**
Write a program that calculates the average marks of students and assigns grades.

**Input Format:**
- Number of students `n`.
- Marks of each student (multiple subjects).

**Output Format:**
- For each student, display average marks and grade.

**Grade Criteria:**
- Average ≥ 90 → Grade A
- Average ≥ 75 → Grade B
- Average ≥ 50 → Grade C
- Otherwise → Fail


**Practice Exercise:**
- Print class topper’s name and marks.
- Show pass percentage of the class.


## 🔹 Project 3: Shopping Cart Billing System

**Problem Statement:**
Create a simple shopping cart system that calculates total bill with discount and tax.

**Input Format:**
- User enters item name and price repeatedly until typing 'done'.
- Optionally enter if user is a student.

**Output Format:**
- Show total before discount.
- Apply discount if total > 1000.
- Apply GST = 5%.
- Extra 5% discount if student.
- Show final bill amount.


**Practice Exercise:**
- Print number of items in the cart.
- Show the average price per item.
- Extend program to store both item names and prices.
