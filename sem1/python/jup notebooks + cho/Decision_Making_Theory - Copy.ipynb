# 🧠 Decision Making in Python

In real life, we make decisions every day (e.g., 'If it rains, take an umbrella').  
In programming, we use **conditional statements** to make such decisions.

---

## 🔹 The `if` Statement
Runs a block of code **only if the condition is True**.

````python
x = 10
if x > 5:
    print('x is greater than 5')
````

## 🔹 The `if-else` Statement
Provides an **alternative** path.

````python
age = 18
if age >= 18:
    print('You can vote!')
else:
    print('You cannot vote yet.')
````

````python
number = int(input('Enter a number: '))

if number % 2 == 0:
    print('The number is Even')
else:
    print('The number is Odd')
````

## 🔹 The `if-elif-else` Statement
Used when there are **multiple conditions**.

````python
marks = 82

if marks >= 90:
    print('Grade: A')
elif marks >= 75:
    print('Grade: B')
elif marks >= 50:
    print('Grade: C')
else:
    print('Grade: Fail')
````

## 🔹 Nested `if`
An `if` inside another `if`.

````python
age = 20
citizen = True

if age >= 18:
    if citizen:
        print('Eligible to vote')
    else:
        print('Not a citizen, cannot vote')
else:
    print('Too young to vote')
````

## 🔹 Match-Case (Python 3.10+)
Similar to **switch-case** in other languages.

````python
choice = 2

match choice:
    case 1:
        print('You chose Option 1')
    case 2:
        print('You chose Option 2')
    case _:
        print('Invalid choice')
````

## ✅ Relational & Logical Operators with Conditions

````python
x = 15
if x > 10 and x < 20:
    print('x is between 10 and 20')

y = 5
if not y == 10:
    print('y is not 10')
````

## ⚠️ Indentation in Python
Python uses indentation instead of `{}` braces.

## 🌤️ Real-Life Example: Temperature Checker

````python
temperature = 32

if temperature > 35:
    print("It's too hot 🌞")
elif temperature >= 20:
    print("The weather is pleasant 🌤️")
else:
    print("It's cold ❄️")
````

# 🏁 Summary

- `if` → check single condition  
- `if-else` → choose between two options  
- `if-elif-else` → handle multiple conditions  
- `nested if` → condition inside condition  
- `match-case` (Python 3.10+) → works like switch-case  
- Use **relational + logical operators** to form conditions  
- Always respect **indentation** in Python
